```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('covid_vaccine_statewise.csv')

if df.empty :
    print("dataframe is empty")
else:
    print("Dataset loaded successfully!")
```

```python
first_dosage = max(df.groupby('State').last(1)['First Dose Administered'])
first_dosage
print("Number of persons state wise vaccinated for first dose in India= {}".format(first_dosage))
```

```python
second_dosage = max(df.groupby('State').last(1)['Second Dose Administered'])
print("Number of persons state wise vaccinated for second dose in India = {}".format(second_dosage))
```

```python
male_dosage = (df.groupby('State').last(1)['Male(Individuals Vaccinated)']).unique()[1]
print("Number of Males vaccinated = {}".format(male_dosage))
```

```python
female_dosage = df.groupby('State').last(1)['Female(Individuals Vaccinated)'].unique()[1]
print("Number of Females vaccinated = {}".format(female_dosage))
```
